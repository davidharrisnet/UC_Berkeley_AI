### Problem 7: Scatterplot of GDP vs. Life Expectancy



`matplotlib` also has a scatterplot method that you can use with `plt.scatter()`.  Use this to create a scatterplot with the GDP column on the x-axis and Life Expectancy on the y-axis.  Add human-readable labels "GDP Per Capita" and "Life Expectancy (years)" respectively and a title "GDP vs. Life Expectancy".



````python
#Your plot here#Your plot hereplt.hist(gap_df['lifeExp'])

x = gap_df['gdpPercap']
y = gap_df['lifeExp']

# Create the scatter plot
plt.scatter(x, y)

# Add labels and a title
plt.title("GDP vs. Life Expectancy")
plt.ylabel("Life Expectancy (years)")
plt.xlabel("GDP Per Capita")


plt.show()

````

### Problem 8: Logarithmic Transformation



Notice the non-linear relationship between GDP and Life Expectancy.  One common transformation is to take something that looks exponential and take a logarithm in order to transform the relationship into a linear one.  Draw a scatterplot like the above, but use `np.log()` to plot the logarithm of GDP vs. Life Expectancy.  Add appropriate x and y labels, and the title "Logarithm of GDP vs. Life Expectancy".



````python
#Your plot here#Your plot here#Your plot hereplt.hist(gap_df['lifeExp'])

x = gap_df['gdpPercap']
y = gap_df['lifeExp']

logx = np.log(x)
logy = np.log(y)

# Create the scatter plot
plt.scatter(x=logx, y=logy)

# Add labels and a title
plt.title("GDP vs. Life Expectancy")
plt.ylabel("Life Expectancy (years)")
plt.xlabel("GDP Per Capita")


plt.show()

````
