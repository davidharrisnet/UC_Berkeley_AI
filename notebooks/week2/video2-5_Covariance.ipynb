```python
import numpy as np
from scipy import stats
```

```python
type(bill_length)
```

```python
bl = bill_length.dropna().to_numpy()
```

```python
stats.zscore(bl,nan_policy='omit') # 'omit' ignores nan and computes the z-score for the rest
```

```python
bl.mean()
```

```python
bl.std()
```

```python
x = 2
def zscore(x,arr):
    return float((x - arr.mean()) / arr.std()) # float converts the np float
```

```python
zscore(x,bl)
```

```python
zscore(bl.mean(),bl)
```

```python
zscore(bl.max(),bl)
```

```python
zscore(bl.min(),bl)
```
