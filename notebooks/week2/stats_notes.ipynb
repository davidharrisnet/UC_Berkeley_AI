## Uniform Distribution

### Uniform 
* all events same probability
* between a and b
* Discrete
    * n possible outcomes
    * E(X) = (a + b)/2
    * Var(X) = (n**2 -1)/12
* Continous
    * E[X] = (a+b)/2
    * Var(X) = (b-a)**2/12
    * Area under the curve is 1

```python
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 10
b = 15
```

* https://docs.scipy.org/doc/scipy/tutorial/stats/discrete.html

```python
u = uniform(loc=10, scale=5) # 10 to 15
```

### pdf
Probability Density Funtion for Uniform Continous Distribution is

f(x) = 1/(b - 1)

```python
float(u.pdf(8))
```

```python
f = float(u.pdf(12)) # cast from np.float(12) to 12
f
```

```python
def pdf(a,b):
    return 1 / ( b - a )

print(pdf(a,b))
```

```python
f = u.pdf([8,12,13, 14, 20])
```

```python
f
```

### Measurements
* pdf f(x) = 1 / ( b - a ) for x in the set [a,b], and 0 otherwise
* mean = (a + b) /
* variance = (b - a)**2 / 12
* std = sqrt(variance)
* cdf - cumulative distributiopn

```python
u.mean()
```

```python
def mean(a,b):
    return (b+a)/2
mean(a,b)
```

```python
u.var()
```

```python
def var(a,b):
    return (b-a)**2/12
var(a,b)
```

```python
u.std()
```

```python
import math
def std(a,b):
  return (b-a)/math.sqrt(12)
std(a,b)
```
