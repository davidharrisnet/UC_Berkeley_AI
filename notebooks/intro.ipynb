# Numpy

### Python arrays

```python
mass = [ 12, 20, 32]
volume = [ 36, 18, 10]
```

```python
try:
  mass/volume
except Exception as e:
    print(e)
```

```python
import numpy as np
```

```python
mass_array = np.array(mass)
volume_array = np.array(volume)
```

```python
density = mass_array/volume_array
```
