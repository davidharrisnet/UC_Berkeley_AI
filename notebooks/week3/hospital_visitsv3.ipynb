```python
# +

df= df[df["Health Region"] != "Out of State"]
```

```python
years = df["Year"].unique()
years
```

```python
#df_inf_count = df_va[df_va["Diagnosed Influenza Count"].apply(lambda x: x.isnumeric())]
df["FlueCount"] = pd.to_numeric(df["Diagnosed Influenza Count"], errors='coerce')
# Or simply:
# df['your_numeric_column'] = df['your_numeric_column'].astype(float)
```

```python
df = df.dropna(subset="FlueCount")
```

```python
df[df["FlueCount"].isna()]
```

```python
df_2025 = df[df["Year"] == 2025]
 
len(df_2025)
```

```python
df_2025_wk = df_2025.groupby(['Week Ending Date', 'Health Region'], as_index=False)['FlueCount'].sum()
```

```python
df_2025_wk
```

```python
fig = px.line(df_2025_wk, x="Week Ending Date", y="FlueCount", color="Health Region")
fig.write_image('images/flu_regions_2025.png')
```
