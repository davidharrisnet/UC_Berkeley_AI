Problem
A coin is flipped three times. Determine a possible random variable that can be assigned to represent the number of heads observed in this experiment.
[openstax](https://openstax.org/books/principles-data-science/pages/3-5-discrete-and-continuous-probability-distributions)

Three coins with two possibilities  2**3 == 8 

 C(N,k) = N! / (k! Ã— (N-k)!)

htt, tht, tth 

three coins -> N == 3
one is heads -> k == 1

C(3,1) = 3!(1! x (3-1)! = 6/(1x2) = 3

```python
from math import factorial
```

```python
def C(N,k):
    return factorial(N) // (factorial(k) * factorial(N-k))
```

```python
C(3,1)
```

For two dice with s sides each, the number of ways to get a sum t:

If t <= s + 1: there are t - 1 ways
If t > s + 1: there are 2s - t + 1 ways

For two d20s summing to 9:

9 <= 21, so: 9 - 1 = 8 ways

```python
import statistics

data = [34.4, 36.4, 37.5, 35.5, 31.3, 36.7]

# Calculate sample variance
sample_variance = statistics.variance(data)
print(f"Sample Variance: {sample_variance}")

# Calculate population variance
population_variance = statistics.pvariance(data)
print(f"Population Variance: {population_variance}")
```
